```python
def extract_from_string(s):
    return ast.literal_eval(re.search('({.+})', s).group(0))
```

```python
def clean_category (category):
    string = category
    string = string.replace("\'", "\"")
    string = string.replace('\\n', ' ')
    string = string.replace('\\r', ' ')
    string = string.replace('\\u', ' ')
    string = string.replace('\\x', ' ')
    string = string.replace('\u200c', ' ')
    
    return string
```

```python
def pre_process(data, is_train):
    data = data.copy()
    data['product_description'] = data['product_description'].apply(lambda s: extract_from_string(s))
    expanded_data = data.drop('product_description', axis=1).join(pd.DataFrame(data['product_description'].values.tolist()))
    
    string_1 = 'دسته بندی'
    string_2 = 'برند'
    data = expanded_data.rename({string_1: 'category', string_2:'brand'}, axis=1)
    
    data['category'] = clean_category(data['category'])
    
    return data
```

```python
train_data = pre_process(train, True)
test_data = pre_process(test, False)
```
